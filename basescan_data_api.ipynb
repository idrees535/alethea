```python
import requests
import pandas as pd

# API details
api_key = 'WRDJ9CNIJW7IT5VG7R71N2I1VSNNJVTAZB'
base_url = 'https://api.basescan.org/api'
contract_address = '0x87a7e67a7eb6BC533d92026f8cb70687eb75C32D'
token_address = '0x97c806e7665d3afd84a8fe1837921403d59f3dcc'  # Replace with the actual contract address of the ALI token

# Construct the URL for the API call for token transfers
url = f"{base_url}?module=account&action=tokentx&contractaddress={token_address}&address={contract_address}&apikey={api_key}"

# Make the API call
response = requests.get(url)
data = response.json()
print(data)

# Check if the response contains data
if data['status'] == '1' and 'result' in data:
    transactions = data['result']
    
    # Create a DataFrame from the transactions data
    df = pd.DataFrame(transactions)
    # Assuming the API response has these fields relevant for token transfers
    df = df[['hash', 'functionName',  'timeStamp', 'from', 'to', 'value']]
    df.rename(columns={
        'hash': 'Transaction Hash',
        'functionName': 'functionName',
        'timeStamp': 'Date Time (UTC)',
        'from': 'From',
        'to': 'To',
        'value': 'Amount'
    }, inplace=True)
    # Convert timestamp to datetime and value to appropriate decimal (depending on token's decimals)
    df['Date Time (UTC)'] = pd.to_datetime(df['Date Time (UTC)'], unit='s')
    df['Amount'] = pd.to_numeric(df['Amount'], errors='coerce') / (10 ** 18)  # Adjust based on token's decimal places
    
    print(df.head())
else:
    print("No data found or error in the API response")
```
